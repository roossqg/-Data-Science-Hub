```python
import pandas as pd
import matplotlib.pyplot as plt


pd.options.display.max_rows=9999
pd.options.display.max_columns=9999
c=pd.read_csv("anime_dataset.csv")

p=pd.Series(c["studios"],index=c["popularity"])
ix=list(c["popularity"])
er=pd.Series(c["popularity"],index=c["studios"])
#print(er)

hash_table={}


for x in c["studios"]:

    if not ',' in x:
        
    
        if x in hash_table:
            hash_table[x][0]+=ix.pop(0)
            hash_table[x][1]+=1

        elif not x in hash_table:
            hash_table[x]=[ix.pop(0),1]



hash_map={}

for x in hash_table.keys():
    if not x in hash_map:
        hash_map[x]=hash_table[x][0]/hash_table[x][1]
    else:


        hash_map[x]+=hash_table[x][0]/hash_table[x][1]


act={f'{x}':hash_map[x] for x in hash_map.keys()}

sss=dict(sorted(act.items(),key=lambda item:item[1]))
ad=pd.Series(sss)
print(ad.head(10))




# 1.Studio for popularity in works
```

```python
import matplotlib.pyplot as plt


x=list(sss.values())
y=[x[2:-3] for x in list(sss.keys())]
cc=[x*12 for x in range(10)]
yy=y[0:10]
xx=x[0:10]

plt.figure(figsize=(19,12))
plt.xlabel("Studio Solo")
plt.ylabel("Popularity")
plt.grid(axis="y",color="r")
plt.bar(yy[::],xx)
plt.title("Anime Studio per Popularity in Solo Works")
plt.show()
```
