```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x=[x+x for x in range(100)]
y=[x*x for x in range(100)]

po=np.diff(y) #--> var:60 between each
pi=np.diff(x) #--> var:1  #--> variance
print(po)
variance=po-pi
print(variance)


plt.plot(y,x,linewidth=0.7,color="b",mec="r") # first element subs
plt.fill_between(x[1:],variance)
#plt.plot(x,ylinewidth=0.7,color="r",mec="b")
plt.xlabel("cars")
plt.ylabel("gasoline")
plt.grid(axis="y",linestyle="--",color="green")
plt.show()
```
