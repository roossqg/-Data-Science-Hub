```python
import tensorflow as tf
from tensorflow import keras
```

## Data

```python

fashion_mnist = keras.datasets.fashion_mnist
(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()

x_train,x_val = x_train[:5000] / 255.0 , x_train[5000:] / 255.0
y_train,y_val = y_train[:5000] / 255.0, y_train[5000:] / 255.0

```

```python
def Ann_Fashion(n_hidden=2,n_neurons=20,learning_rate=3e-2,
            activation='relu',optimizer='SGD',
            loss='scc',metrics=['accuracy'],input_shape=[4],aux_output='yes'):

    model = keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=input_shape))
    for n_hidden in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons,activation=activation))
    model.add(keras.layers.Dense(10,activation='softmax'))
    model.compile(loss=loss,metrics=metrics,optimizer=optimizer)

```

```python
from scikeras.wrappers import KerasRegressor
classifier_fashion_mnist = KerasRegressor(model=Ann_Fashion)
```

## Training Pipeline

```python
from scipy.stats import reciprocal
from scikit import RandomizedSearchCV

parans_dist={
        n_hidden=[2,4,6],n_neurons=[200,300,400],learning_rate=[3e-2,2e-2,1e-2],
        activation='relu',optimizer='SGD',
        loss='scc',metrics=['accuracy'],input_shape=[28,28],aux_output='yes'
    
}

rns = RandomizedSearchCV(classifier_fashion_mnist,parans_dist,n_iter=10,cv=3)
rns.fit(x_train,y_train,epochs=20,validation_data=(x_val,y_val),
        callbacks=[keras.EarlyStopping=(patience=10)])
```

```python
print(rns.best_model,rns.best_parans,rns.best_score)
```
